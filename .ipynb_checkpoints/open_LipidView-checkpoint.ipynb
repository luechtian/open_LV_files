```python
########################################################################################
import os
import glob
import numpy
import pandas as pd

########################################################################################
def open_LVfile(LVfile):
    """
    This function opens a single text file, removes redundant columns and renames the columns.
    this function returns cleaner table.
    """
    col_names = list(pd.read_table(LVfile, skiprows=1, usecols = lambda column : 'Unnamed' not in column))
    lipid_data = pd.read_table(LVfile, skiprows=2, index_col=0, usecols = lambda column : column not in ['Sample ID', 'PIS m/z', 'Polarity', 'View Type', '(ScanName)'])
    lipid_data.columns = col_names[1:]
    return lipid_data

########################################################################################
LV_file = os.path.join('data', '200131_XtrIL_AMV_SM.txt')
open_LVfile(LV_file).head(4)

########################################################################################
file_location = os.path.join('data', '*.txt')
filenames = glob.glob(file_location)

data = []
for i in range(0,len(filenames)):
    dat = open_LVfile(filenames[i])
    data.append(dat)

comb_data = []
for i in range(0,1):
    comb_data = data[i]
    for i2 in range(1, 15):
        comb_data = pd.concat([comb_data, data[i2]], axis=0)

comb_data.to_csv('comb_data.csv')
comb_data
```

```python
def check_lipidclass(dataset, column):
    """
    
    """
    lipidclass = dataset[dataset[column].notnull()]
    
    return lipidclass
```

```python
df = check_lipidclass(comb_data,'kon A')
df = df.fillna(0)      #if the blanks are nan will need this line first
df['Kon A'] = df['kon A'] + df['Kon A']

df.drop('kon A', axis=1)
```
